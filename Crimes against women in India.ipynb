# Data Preparation and Cleaning

```python
df.isna().sum()
```

There are 718 districts in India, in total, which means there is messy or false datas in a huge amount, in this case, we better drop the column "District" and also "Unnamed: 0", as it is of no use, in our data analysis process

```python
df.drop(['DISTRICT', 'Unnamed: 0'], axis = 1, inplace=True)
```

```python
df['STATE/UT'].unique()
```

```python
def remove_uppercase(r):
    r = r['STATE/UT'].strip()
    r = r.upper()
    return r
df['STATE/UT'] = df.apply(remove_uppercase, axis=1)

#Now use replace function to replace the other type of repeated datas as dicussed above
df['STATE/UT'].replace("A&N ISLANDS", "A & N ISLANDS", inplace = True)
df['STATE/UT'].replace("D&N HAVELI", "D & N HAVELI", inplace = True)
df['STATE/UT'].replace("DELHI UT", "DELHI", inplace = True)
```

```python
len(df['STATE/UT'].unique())
```

```python
year_df=df.groupby('Year')[['Rape','Kidnapping and Abduction','Dowry Deaths','Assault on women with intent to outrage her modesty','Insult to modesty of Women','Cruelty by Husband or his Relatives','Importation of Girls']].sum()
```

```python
# Create a dataframe containing 10 highest reported deaths caused by Dowry cases in India, in the span of year 2001-2014.

max_dowry_death_cases = df.sort_values('Dowry Deaths', ascending = False).head(10)
max_dowry_death_cases
```

From our analysis, we observe that the highest reported dowry death was in Uttar Pradesh in 2014, with number of reports being 2469.
One more here to be observed as well is that, Uttar Pradesh is the ONLY state that appears in this list

```python
# Create a dataframe containing 10 highest reported Domestic Violance cases in India, in the span of year 2001-2014.
max_domestic_violance_cases = df.sort_values('Cruelty by Husband or his Relatives', ascending = False).head(10)
max_domestic_violance_cases
```

According to our analysis, we see the maximum cases of Domestic Violance cases came from West Bengal in the year 2014, with number of cases reported being 23278.

```python
# Create a dataframe containing 10 highest reported Importation cases in India, in the span of year 2001-2014.

max_importation_case = df.sort_values('Importation of Girls', ascending = False).head(10)
max_importation_case
```

```python
# Find out the total number of cases, in span of 2001-2014 under each category, state wise.
counts_df = df.groupby('STATE/UT')[['Rape', 'Kidnapping and Abduction', 'Dowry Deaths','Cruelty by Husband or his Relatives', 'Importation of Girls']].sum()
counts_df
```

```python
# Find out the top 5 states, where maximum numbers of cases has been reported in TOTAL in span of 2001-2014, each category wise.

counts_df.sort_values(by = 'Rape', ascending = False).head(5)
```

Madhya Pradesh has reported the highest number of rape cases in TOTAL in span of 2001-2014, where UP, Maharastra, West Bengal and Rajasthan follows the list.

```python
counts_df.sort_values(by = 'Kidnapping and Abduction', ascending = False).head(5)
```

Uttar Pradesh has reported the highest number of cses under "Kidnapping and Abduction" in TOTAL in span of 2001-2014, where Rajasthan, Assam, West Bengal and Bihar follows the list.

```python
counts_df.sort_values(by = 'Dowry Deaths', ascending = False).head(5)
```

Uttar Pradesh has reported the highest number of Deaths caused by Dowry cases in TOTAL in span of 2001-2014, where Bihar, Madhya Pradesh, Andhra Pradesh and West Bengal follows the list

```python
counts_df.sort_values(by = 'Cruelty by Husband or his Relatives', ascending = False).head(5)
```

West Bengal has reported the highest number of cases of Domestic Violence in TOTAL in span of 2001-2014, where Andhra Pradesh, Rajsthan, Uttar Pradesh and Maharashtra follows the list.

```python
counts_df.sort_values(by = 'Importation of Girls', ascending = False).head(5)
```

Bihar has reported the highest number of Importation of girl's cases in TOTAL in span of 2001-2014, where Jharkhand, West Bengal, Madhya Pradesh and Karnataka follows the list.
